```python
import scanpy as sc
import pandas as pd
from pathlib import Path
from google.colab import drive
```

```python
drive.mount("/content/drive")
```

```python
DATA = Path("/content/drive/My Drive/ecbm_4060")
```

```python
melanoma_df = pd.read_csv(DATA / "GSE72056_melanoma_single_cell_revised_v2.txt", sep="\t")
# transpose such that the data direction conforms with scanpy defaults
melanoma_df = melanoma_df.T
# rename columns as first row
melanoma_df.columns = melanoma_df.iloc[0, :]
melanoma_df = melanoma_df.iloc[1:, :]
# remove duplicate genes and NA genes
melanoma_df = melanoma_df.loc[:,~melanoma_df.columns.duplicated()]
melanoma_df = melanoma_df.loc[:,~melanoma_df.columns.isna()]
melanoma_df
```

```python
# prepare observations for removal
tumor = melanoma_df["tumor"]
malignant = melanoma_df["malignant(1=no,2=yes,0=unresolved)"]
non_malignant_cell_type = melanoma_df["non-malignant cell type (1=T,2=B,3=Macro.4=Endo.,5=CAF;6=NK)"]
# remove first 3 rows as observations
melanoma_df = melanoma_df.iloc[:, 3:]
melanoma_df
```

```python
melanoma_adata = sc.AnnData(melanoma_df,
                            melanoma_df.index.to_frame(name="cell"),
                            melanoma_df.columns.to_frame(name="gene"))

# reapply observations, note that the datatype changes to suit casting
melanoma_adata.obs["tumor"] = tumor.astype("int64")
melanoma_adata.obs["malignant"] = malignant.astype("category")
melanoma_adata.obs["non_malignant_cell_type"] = non_malignant_cell_type.astype("category")
```

```python
# export to use in analysis
melanoma_adata.write(filename=DATA / "GSE72056.h5ad")
```
