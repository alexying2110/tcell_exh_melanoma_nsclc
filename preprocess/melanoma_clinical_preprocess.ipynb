```python
import scanpy as sc
import pandas as pd
from pathlib import Path
from google.colab import drive
```

```python
drive.mount("/content/drive")
```

```python
DATA = Path("/content/drive/My Drive/ecbm_4060")
```

```python
a_df = pd.read_csv(DATA / "GSE189125_5prime_scRNAseq_seqbatchA_expression.txt", sep="\t", index_col=0)
# transpose such that the data direction conforms with scanpy defaults
a_df = a_df.T
a_df
```

```python
a_meta = pd.read_csv(DATA / "GSE189125_5prime_scRNAseq_seqbatchA_metadata.txt", sep="\t", index_col=0)
a_meta
```

```python
a_adata = sc.AnnData(a_df,
                     a_df.index.to_frame(name="cell"),
                     a_df.columns.to_frame(name="gene"))


# reapply observations, note that the datatype changes to suit casting
a_adata.obs["ptid"] = a_meta.loc[a_df.index.values]["patient ID"].values
a_adata.obs["cell_type"] = a_meta.loc[a_df.index.values]["unsupervised cell cluster"].str.split(" ").str[0].values
```

```python
# export to use in analysis
a_adata.write(filename=DATA / "GSE189125A.h5ad")
```
